## Data Cleaning

```python
# Each cuisine converts in to separate row.

df['cuisines'] = df['cuisines'].str.split(',')
df = df.explode('cuisines')
df['cuisines'] = df['cuisines'].str.strip()
```

```python
df['cuisines'].unique()
```

```python
# removing non-cuisine items

non_cuisine = [
    "Cafe","Coffee","Tea","Bubble Tea","Juices","Drinks Only",
    "Beverages","Cafe Food","Coffee and Tea"
]

df = df[~df['cuisines'].isin(non_cuisine)]
```

```python
# handling missing values

df = df.dropna(subset=['cuisines'])
```

```python
# standardize similar cuisines

df['cuisines'] = df['cuisines'].replace({
    "Awadhi":"North Indian",
    "Hyderabadi":"North Indian",
    "Lucknowi":"North Indian",
    "Chettinad":"South Indian",
    "Andhra":"South Indian",
    "Kerala":"South Indian",
    "Cantonese":"Chinese",
    "Pan Asian":"Asian"
})
```

```python
df[(df['zipcode'].isnull())& (df['locality'].notnull())]['locality'].value_counts().reset_index()
```

#### It shows the count of empty cell value which has repetitive name in locality

```python
df['establishment'] = df['establishment'].str.strip("[]")
df['establishment'] = df['establishment'].str.replace("'", "")

df['establishment'] = df['establishment'].replace('', pd.NA)
```

```python
df['establishment'].unique()
```

```python
top_cuisine=df.groupby('city')['cuisines'].value_counts().reset_index()
top_cuisine
```

```python
# This shows the city's top cuisine

city_top_cuisine = top_cuisine.sort_values('count', ascending=False).groupby('city').head(1)
city_top_cuisine
```

```python
city_cuisine = city_top_cuisine.head(20)

plt.figure(figsize=(25,8))
ax=sns.barplot(data=city_cuisine, x="city", y="count", hue = "cuisines", palette = "Set2")

ax.bar_label(ax.containers[0])

for container in ax.containers:
    ax.bar_label(container)

plt.xticks(rotation=45)

plt.title("Top Cuisine in 20 Cities")
plt.ylabel("Number of Restaurants")
plt.xlabel("City")
```

#### North Indian cuisine appears as the most dominant cuisine across most major cities,

### restaurant order count vs their rating city wise

```python
result = df.groupby(['city', 'name']).agg(
    order_count=('name', 'count'),
    avg_rating=('aggregate_rating', 'mean')
).sort_values('order_count', ascending =False).reset_index()

result
```

```python
most_ordered = result.loc[
    result.groupby('city')['order_count'].idxmax()
]
most_ordered
```

```python
highest_rated = result.loc[
    result.groupby('city')['avg_rating'].idxmax()
]
highest_rated
```

```python
final = pd.merge(
    most_ordered[['city', 'name', 'order_count']],
    highest_rated[['city', 'name', 'avg_rating']],
    on='city',
    suffixes=('_most_ordered', '_highest_rated')
)
ac=final.head(20)
ac
```

#### This comparison highlights the gap between popularity and customer satisfaction at a city level.
